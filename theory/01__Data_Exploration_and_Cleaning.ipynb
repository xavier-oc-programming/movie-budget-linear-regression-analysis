### **Challenge 2 — Convert Budget and Revenue Columns**

Columns to convert:

* `USD_Production_Budget`
* `USD_Worldwide_Gross`
* `USD_Domestic_Gross`

Characters to remove:

* Dollar signs (`$`)
* Commas (`,`)

---

### **Solution — Challenge 2**

```python
chars_to_remove = [',', '$']
columns_to_clean = [
    'USD_Production_Budget',
    'USD_Worldwide_Gross',
    'USD_Domestic_Gross'
]

for col in columns_to_clean:
    for char in chars_to_remove:
        data[col] = data[col].astype(str).str.replace(char, "")
    data[col] = pd.to_numeric(data[col])
```

Verification:

```python
data.head()
```

All budget and revenue columns are now numeric.
