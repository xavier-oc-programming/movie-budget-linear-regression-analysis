## **4)—Seaborn Data Visualisation: Bubble Charts**

### **Overview**

In this section we use **Seaborn** as a high-level data visualisation library built on top of **Matplotlib**.
We start with a basic scatter plot, then customise it using Matplotlib, and finally transform it into a **bubble chart** by encoding additional variables as colour and size.

This allows us to visualise **three dimensions at once** in a single chart.

---

### **Import Seaborn**

By convention, Seaborn is imported using the alias `sns`.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

---

### **Seaborn Scatter Plot**

A Seaborn scatter plot only requires:

* a DataFrame
* the column names for the x and y axes

```python
sns.scatterplot(
    data=data_clean,
    x='USD_Production_Budget',
    y='USD_Worldwide_Gross'
)
```

### **Adjust Figure Size Using Matplotlib**

Because Seaborn is built on top of Matplotlib, we can control the figure size before plotting:

```python
plt.figure(figsize=(8, 4), dpi=200)

sns.scatterplot(
    data=data_clean,
    x='USD_Production_Budget',
    y='USD_Worldwide_Gross'
)

plt.show()
```

This improves readability and resolution without changing the data itself.

---

### **Customising Axes with the Axes Object**

`sns.scatterplot()` returns a Matplotlib **Axes** object.
This allows us to directly control axis limits and labels.

```python
plt.figure(figsize=(8, 4), dpi=200)

ax = sns.scatterplot(
    data=data_clean,
    x='USD_Production_Budget',
    y='USD_Worldwide_Gross'
)

ax.set(
    ylim=(0, 3000000000),
    xlim=(0, 450000000),
    ylabel='Revenue in $ billions',
    xlabel='Budget in $100 millions'
)

plt.show()
```

Here we:

* Limit the visible range of the data
* Rename axis labels to more meaningful units

---

### **From Scatter Plot to Bubble Chart**

Seaborn makes it easy to encode extra information using:

* `hue` for colour
* `size` for point size

```python
plt.figure(figsize=(8, 4), dpi=200)

ax = sns.scatterplot(
    data=data_clean,
    x='USD_Production_Budget',
    y='USD_Worldwide_Gross',
    hue='USD_Worldwide_Gross',
    size='USD_Worldwide_Gross'
)

ax.set(
    ylim=(0, 3000000000),
    xlim=(0, 450000000),
    ylabel='Revenue in $ billions',
    xlabel='Budget in $100 millions'
)

plt.show()
```

Now the chart represents **three dimensions**:

* Budget (x-axis)
* Revenue (y-axis)
* Revenue again via colour and bubble size

Higher-grossing movies appear larger and darker.

---

### **Styling a Single Chart with `with sns.axes_style()`**

To apply styling to **only one chart**, we use Python’s `with` keyword:

```python
plt.figure(figsize=(8, 4), dpi=200)

with sns.axes_style('darkgrid'):
    ax = sns.scatterplot(
        data=data_clean,
        x='USD_Production_Budget',
        y='USD_Worldwide_Gross',
        hue='USD_Worldwide_Gross',
        size='USD_Worldwide_Gross'
    )

    ax.set(
        ylim=(0, 3000000000),
        xlim=(0, 450000000),
        ylabel='Revenue in $ billions',
        xlabel='Budget in $100 millions'
    )

plt.show()
```

This styling applies **only inside the block**, not to the entire notebook.

Other built-in Seaborn themes include:

* `whitegrid`
* `dark`
* `ticks`

### **Challenge: Movie Budgets Over Time**

The goal is to visualise how movie budgets have evolved over time while still encoding worldwide revenue.

This chart represents:

* Release year on the x-axis
* Production budget on the y-axis
* Worldwide gross via colour and size

```python
plt.figure(figsize=(8, 4), dpi=200)

with sns.axes_style("darkgrid"):
    ax = sns.scatterplot(
        data=data_clean,
        x='Release_Date',
        y='USD_Production_Budget',
        hue='USD_Worldwide_Gross',
        size='USD_Worldwide_Gross'
    )

    ax.set(
        ylim=(0, 450000000),
        xlim=(data_clean.Release_Date.min(),
              data_clean.Release_Date.max()),
        xlabel='Year',
        ylabel='Budget in $100 millions'
    )

plt.show()
```

---

### **Analysis**

The chart shows a clear structural change in the film industry:

* Early decades feature few films with relatively small budgets
* From the late 1970s onward, budgets begin to rise sharply
* From around 2000 onward, releases become extremely dense
* High-budget, high-revenue films stand out clearly as large, dark bubbles

This visualisation makes both **growth in budgets** and **growth in film production volume** immediately obvious.
