## **7— Use scikit-learn to Run Your Own Regression**

### **Overview**

In this section, we move beyond visual regression lines and **run an explicit linear regression model** using **scikit-learn**.
We fit a **univariate linear regression** where:

* **Feature (X)** → Movie production budget
* **Target (y)** → Worldwide revenue

This allows us to:

* Extract the **intercept** and **slope**
* Quantify **how much revenue increases per dollar of budget**
* Measure **how well the model fits the data** using **R²**
* Make **numerical predictions**

---

### **Linear Regression Model Structure**

The model follows this equation:

[
\widehat{\text{Revenue}} = \theta_0 + \theta_1 \times \text{Budget}
]

* **θ₀ (intercept)** → Estimated revenue when budget = 0
* **θ₁ (slope)** → Extra revenue per additional $1 of budget

---

### **Import Statements**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

---

### **Create the Regression Object**

```python
regression = LinearRegression()
```

This object will:

* Learn the best-fit line
* Store the intercept and slope after training

---

### **Define Features (X) and Target (y)**

scikit-learn expects **DataFrames**, not Series.

```python
# Feature (explanatory variable)
X = pd.DataFrame(new_films, columns=['USD_Production_Budget'])

# Target (response variable)
y = pd.DataFrame(new_films, columns=['USD_Worldwide_Gross'])
```

---

### **Fit the Model**

```python
regression.fit(X, y)
```

This computes:

* The optimal intercept (θ₀)
* The optimal slope (θ₁)

---

### **Extract Model Parameters**

```python
# Intercept (theta zero)
regression.intercept_

# Slope (theta one)
regression.coef_
```

**Results (approximate):**

* **Intercept** ≈ −8.65 million
* **Slope** ≈ 3.12

---

### **Interpretation of Parameters**

#### **Intercept**

* Predicts revenue at **$0 budget**
* Negative value is **not realistic**
* Happens because the model is forced to be linear
* Interpretation is most reliable **where data is dense**, not at extremes

#### **Slope**

* For every **$1 increase in budget**, revenue increases by **~$3.1**
* Indicates a **strong positive relationship** between budget and revenue

---

### **R-Squared: Goodness of Fit**

```python
regression.score(X, y)
```

**Result:**

* **R² ≈ 0.558**

**Interpretation:**

* About **56% of the variance in revenue** is explained by budget alone
* Very strong result for a **single-feature model**
* Real-world data is complex — anything above ~50% is considered solid
