## **5)—Floor Division: A Trick to Convert Years to Decades**

### **Introduction**

To better analyse long-term trends in the film industry, it is useful to group movies by **decade** instead of individual years.
This section introduces a simple but powerful technique using **floor division** to convert release years into decade labels.

---

### **Objective**

Add a new column called `Decade` to `data_clean` such that:

* A movie released in **1992** or **1999** is assigned **1990**
* A movie released in **2005** is assigned **2000**

---

### **Challenge**

Create a decade column in `data_clean` using the `Release_Date` column.

Steps required:

1. Create a `DatetimeIndex` from `Release_Date`
2. Extract the release year
3. Use floor division (`//`) to convert years into decades
4. Store the result in a new `Decade` column

---

### **Solution — Convert Years to Decades**

#### **Step 1: Create a DatetimeIndex and extract years**

```python
dt_index = pd.DatetimeIndex(data_clean.Release_Date)
years = dt_index.year
```

---

#### **Step 2: Use floor division**

Floor division (`//`) performs integer division and rounds down:

```python
5 / 2    # 2.5
5 // 2   # 2
```

Applied to years:

```python
1999 // 10 * 10
```

Result:

```text
1990
```

---

#### **Step 3: Add the Decade column**

```python
decades = years // 10 * 10
data_clean['Decade'] = decades
```

The dataset now includes a `Decade` column suitable for grouped analysis.

---

### **Challenge — Split Old and New Films**

Create two new DataFrames:

* `old_films`: films released **before 1970** (up to and including 1969)
* `new_films`: films released **from 1970 onwards**

---

### **Solution — Separate Films by Era**

```python
old_films = data_clean[data_clean.Decade <= 1960]
new_films = data_clean[data_clean.Decade > 1960]
```

Explanation:

* Using `1960` as the cutoff includes all films released up to **1969**
* The dataset shows far fewer films before 1970 than after

---

---

### **Key Takeaways**

* Floor division (`//`) is an elegant way to group continuous values into bins
* Converting years into decades simplifies long-term trend analysis
* The modern film industry (post-1970) dominates the dataset in both volume and budget size
* Early cinema operated with far fewer releases and significantly lower budgets
