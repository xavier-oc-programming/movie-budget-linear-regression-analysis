## **3)—Filter on Multiple Conditions: International Films**

### **Introduction**

Up to this point, subsets of the DataFrame have been created using **single conditions**.
In this section, we filter data based on **multiple conditions simultaneously** to identify films that:

* Made money **internationally**
* Earned **zero revenue in the United States**

---

### **Filtering with Multiple Conditions**

We want films where:

* `USD_Domestic_Gross == 0`
* `USD_Worldwide_Gross != 0`

This can be done using the `.loc[]` indexer combined with the **bitwise AND operator (`&`)**.

```python
international_releases = data.loc[
    (data.USD_Domestic_Gross == 0) &
    (data.USD_Worldwide_Gross != 0)
]

print(f'Number of international releases: {len(international_releases)}')
international_releases.head()
```

---

### **Why This Works**

* Pandas is built on top of NumPy
* NumPy uses **bitwise operators** (`&`, `|`) for element-wise logical comparisons
* Parentheses are required because the bitwise operator has higher precedence than comparison operators

Example of element-wise comparison:

```python
bool_list1 = [True, True, False, False]
bool_list2 = [False, True, True, False]

np.array(bool_list1) & np.array(bool_list2)
```

Result:

```text
[False  True False False]
```

---

### **Alternative Method — Using `.query()`**

The same filter can be created using the Pandas `.query()` method.
In this case, the entire condition is written as a string and uses the `and` keyword.

```python
international_releases = data.query(
    'USD_Domestic_Gross == 0 and USD_Worldwide_Gross != 0'
)

print(f'Number of international releases: {len(international_releases)}')
international_releases.tail()
```

The result is **identical** to the `.loc[]` approach.

### **Unreleased Films**

Some films show zero revenue because they **had not yet been released** at the time the dataset was collected.

The data collection date is:

```python
scrape_date = pd.Timestamp('2018-5-1')
```

Identify unreleased films:

```python
future_releases = data[data.Release_Date >= scrape_date]

print(f'Number of unreleased movies: {len(future_releases)}')
future_releases
```

Results:

* **7 unreleased films** are present in the dataset
* These films have zero revenue because they had not yet reached cinemas

---

### **Removing Unreleased Films**

Create a new DataFrame that excludes unreleased films.

```python
data_clean = data.drop(future_releases.index)
```

From this point forward, all analysis is performed using **`data_clean`**.

---

### **Bonus Challenge — Films That Lost Money**

Identify films where **production budget exceeded worldwide revenue**.

Using `.loc[]`:

```python
money_losing = data_clean.loc[
    data_clean.USD_Production_Budget > data_clean.USD_Worldwide_Gross
]

len(money_losing) / len(data_clean)
```

Using `.query()`:

```python
money_losing = data_clean.query(
    'USD_Production_Budget > USD_Worldwide_Gross'
)

money_losing.shape[0] / data_clean.shape[0]
```

---

### **Key Result**

* **37.2% of films** did not recoup their production budget at the box office

* The financial risk inherent in film production
* Why understanding revenue distributions is essential before modelling relationships with regression
