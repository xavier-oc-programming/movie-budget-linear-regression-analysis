## **8— Learning Points & Summary**

### **Overview**

This lesson brought together **data cleaning, filtering, visualisation, and regression analysis** into a single coherent workflow. The focus was not only on *how* to perform each step, but also on *how to judge whether the results actually make sense*.

---

### **Key Learning Points**

* **Cleaning multiple columns efficiently**

  * Used **nested loops** to remove unwanted characters across several DataFrame columns.
  * Reinforced the idea of writing reusable, scalable data-cleaning logic.

* **Filtering DataFrames with multiple conditions**

  * Applied conditional filtering using:

    * **`.loc[]`** with boolean logic and bitwise operators (`&`)
    * **`.query()`** with readable string-based conditions
  * Compared readability and use cases for each approach.

* **Creating bubble charts with Seaborn**

  * Extended scatter plots using:

    * `hue` to encode a third variable (colour)
    * `size` to encode magnitude (bubble size)
  * Visualised multiple dimensions simultaneously.

* **Styling visualisations**

  * Used Seaborn’s **built-in themes** (`darkgrid`, `whitegrid`, `ticks`)
  * Dropped into the **Matplotlib layer** to:

    * Set axis limits
    * Rename axis labels
    * Control figure size and DPI

* **Converting years into decades**

  * Used **floor division (`//`)** to convert release years into decade bins.
  * Added derived columns to support time-based grouping and comparison.

* **Overlaying linear regressions with Seaborn**

  * Used **`regplot()`** to:

    * Plot scatter points
    * Overlay regression lines
    * Visualise confidence intervals
  * Compared regression quality across different subsets of data.

* **Evaluating regression quality**

  * Interpreted regression results using:

    * Visual alignment of data points
    * Width of confidence intervals
    * **R-squared (R²)** as a quantitative goodness-of-fit metric
  * Learned to recognise when a regression is **useful** versus **misleading**.

* **Running regressions with scikit-learn**

  * Built linear regression models using **`LinearRegression`**
  * Extracted:

    * Intercept
    * Slope (coefficient)
    * R-squared score
  * Used the trained model to make **numerical predictions**.

---

### **Big Picture Takeaway**

This lesson emphasised that **data science is not just about running code**:

* Visualisations must be interpreted critically.
* Models must be evaluated, not trusted blindly.
* A simple model can be powerful — but only within the context where its assumptions hold.

Understanding *when* a model works is just as important as knowing *how* to build it.
