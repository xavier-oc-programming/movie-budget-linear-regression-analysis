## **0)—Day 78 Goals: what you will make by the end of the day**

### **Overview**

In this lesson, we analyse **movie production budgets and box-office revenue** to determine whether higher budgets tend to lead to higher earnings.

The dataset is used to practise:

* Exploratory data analysis
* Statistical data visualisation
* Linear regression modelling

The objective is to **understand relationships in real-world data**, not just to create plots.

---

### **Main Question**

Do higher film budgets lead to higher box-office revenue?

Put differently:

* Does spending more money on a movie increase the likelihood of making more money?
* Is there a measurable relationship between **production cost** and **revenue**?

---

### **Tools Introduced**

* **Seaborn**

  * A statistical data visualisation library
  * Built on top of Matplotlib
  * Especially useful for regression and distribution plots

* **scikit-learn**

  * A machine learning library
  * Used here to run and interpret linear regression models
  * Helps quantify relationships between variables

---

### **What You Will Learn Today**

* How to visualise relationships between variables using Seaborn
* How to run and interpret a linear regression with scikit-learn
* How to plot regression lines on scatter plots
* How to add a third dimension to scatter plots using bubble charts
* How to use floor division (`//`) to convert years into decades for analysis

---

### **End Goal**

By the end of the day, you will have:

* Explored and cleaned a real-world dataset
* Built visual intuition using Seaborn
* Run and interpreted a linear regression model
* Gained insight into how budget relates to revenue in the film industry
