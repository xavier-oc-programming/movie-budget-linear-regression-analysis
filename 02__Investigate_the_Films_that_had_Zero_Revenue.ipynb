### **Challenge 2 — No Domestic Revenue**

How many films grossed **$0 domestically** (United States)?
Which high-budget films reported no domestic revenue?

---

### **Solution — Challenge 2**

```python
zero_domestic = data[data.USD_Domestic_Gross == 0]

print(f'Number of films that grossed $0 domestically {len(zero_domestic)}')

zero_domestic.sort_values('USD_Production_Budget', ascending=False)
```

Results:

* **512 films** grossed $0 domestically
* Many high-budget films were released **after the dataset cutoff date**
* Zero domestic revenue does not necessarily indicate failure

---

### **Challenge 3 — No Worldwide Revenue**

How many films grossed **$0 worldwide**?
Which high-budget films had no international revenue?

---

### **Solution — Challenge 3**

```python
zero_worldwide = data[data.USD_Worldwide_Gross == 0]

print(f'Number of films that grossed $0 worldwide {len(zero_worldwide)}')

zero_worldwide.sort_values('USD_Production_Budget', ascending=False)
```

Results:

* **357 films** grossed $0 worldwide
* Fewer than the domestic-zero films
* Some films earned money internationally but were never screened in the United States
