## **6)—Plotting Linear Regressions with Seaborn**

### **Goal**

Visualise and interpret the relationship between **movie production budget** and **worldwide revenue** using **linear regression**, and compare how this relationship differs between **older films** and **newer films**.

---

### **Using `regplot()` for Linear Regression**

Seaborn provides the `regplot()` function, which:

* Draws a **scatter plot**
* Fits and overlays a **linear regression line**
* Displays a **confidence interval** automatically

#### **Regression for Older Films**

```python
sns.regplot(
    data=old_films,
    x='USD_Production_Budget',
    y='USD_Worldwide_Gross'
)
```

This creates:

* Scatter points for each film
* A regression line estimating the budget–revenue relationship
* A shaded confidence interval around the line

---

### **Styling the Regression Plot**

We can drop into the **Matplotlib layer** to customise transparency, colours, and layout.

```python
plt.figure(figsize=(8, 4), dpi=200)

with sns.axes_style("whitegrid"):
    sns.regplot(
        data=old_films,
        x='USD_Production_Budget',
        y='USD_Worldwide_Gross',
        scatter_kws={'alpha': 0.4},
        line_kws={'color': 'black'}
    )
```

#### **Key Styling Parameters**

* `scatter_kws`: controls scatter appearance (e.g. transparency)
* `line_kws`: controls regression line styling
* `axes_style()`: applies a theme to a single chart

---

### **Interpreting the Regression (Older Films)**

* A very **weak relationship** between budget and revenue
* Many low-budget films outperform expectations
* Several points lie far from the regression line
* The line does **not explain the data well**

A clear outlier appears at the far right (e.g. *Cleopatra*), which heavily influences the regression.

---

### **Challenge: Regression for Newer Films**

Repeat the regression using the **`new_films`** DataFrame and apply the following styling:

* Theme: `'darkgrid'`
* Scatter colour: `#2f4b7c` (dark blue)
* Regression line colour: `#ff7c43` (orange)
* No negative axis values
* Axis labels:

  * **Revenue in $ billions**
  * **Budget in $100 millions**

---

### **Solution: Regression for Newer Films**

```python
plt.figure(figsize=(8, 4), dpi=200)

with sns.axes_style('darkgrid'):
    ax = sns.regplot(
        data=new_films,
        x='USD_Production_Budget',
        y='USD_Worldwide_Gross',
        color='#2f4b7c',
        scatter_kws={'alpha': 0.3},
        line_kws={'color': '#ff7c43'}
    )

    ax.set(
        ylim=(0, 3000000000),
        xlim=(0, 450000000),
        ylabel='Revenue in $ billions',
        xlabel='Budget in $100 millions'
    )
```

---

### **Interpreting the Regression (Newer Films)**

* **Much stronger linear relationship** than with older films
* Data points cluster closer to the regression line
* Confidence interval is noticeably **narrower**
* Indicates budgets explain revenue **far better** in modern cinema

📌 A film with a **$150 million budget** is predicted to earn **just under $500 million worldwide** according to the regression line.

---

### **Key Takeaways**

* Linear regression fits **newer films** significantly better than older ones
* Budget has become a **stronger predictor** of revenue over time
* Visual regression helps assess:

  * Trend strength
  * Outliers
  * Predictive usefulness
